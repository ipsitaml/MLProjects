# Q8 Split the pima_df into training and test set in the ratio of 70:30 (Training:Test).</b>


```python
X = pima_df.drop("class" , axis=1)
Y = pima_df["class"]

```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30 , random_state=1)
```

```python
print ("{0:2.2f}% in training data".format((len(X_train)/len(pima_df.index))*100))
print ("{0:2.2f}% in test data".format((len(X_test)/len(pima_df.index))*100))
```

```python
from sklearn.metrics import accuracy_score
```

# Q9 Create the decision tree model using “entropy” method of reducing the entropy and fit it to training data.</b>



```python
from sklearn.tree import DecisionTreeClassifier
DTclassifier = DecisionTreeClassifier (max_depth=5, random_state = 0, criterion="entropy")
DTclassifier.fit(X_train, Y_train)
predicted_labels = DTclassifier.predict(X_test)
print(accuracy_score(Y_test, predicted_labels))
```

# Q10 Print the accuracy of the model & print the confusion matrix 

```python
print(accuracy_score(Y_test, predicted_labels))
```

```python
print(metrics.confusion_matrix(Y_test, predicted_labels))
```

# Next steps are optional and will not be graded.

Voting is one of the simplest ways of combining the predictions from multiple machine learning algorithms.

It works by first creating two or more standalone models from your training dataset. A Voting Classifier can then be used to wrap your models and average the predictions of the sub-models when asked to make predictions for new data.

Apply Voting Classfier on the given dataset and state your insights.

```python
from sklearn.ensemble import VotingClassifier

clf1 = AdaBoostClassifier(n_estimators=50 , random_state=1)
clf2 = BaggingClassifier( n_estimators=50 , random_state=1)
clf3 = RandomForestClassifier(n_estimators=50 , max_depth=5 ,random_state= 1)
clf4 = GradientBoostingClassifier(n_estimators=50 ,random_state=1)

eclf = VotingClassifier(estimators=[('ab', clf1),('bc' , clf2),('rfc' ,clf3 ),('gbc', clf4)])
eclf.fit(X_train , Y_train)
predicted_labels= eclf.predict(X_test)
print(accuracy_score(Y_test, predicted_labels))
```
